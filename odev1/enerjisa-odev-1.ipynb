#soru 2

```python
df = pd.read_csv("data.csv")
df.describe()
```

```python
print("Number Of Outlier value")
dfB = df[df["diagnosis"] == "B"]
for column in dfB.columns[2:]:
    desc = dfB[column].describe()
    Q1 = desc[4]
    Q3 = desc[6]
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"{column}: {dfB[(dfB[column] < lower_bound) | (dfB[column] > upper_bound)].radius_mean.size}", desc[2])
```

```python
print("Number Of Outlier value")
dfM = df[df["diagnosis"] == "M"]
for column in dfM.columns[2:]:
    desc = dfM[column].describe()
    Q1 = desc[4]
    Q3 = desc[6]
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"{column}: {dfM[(dfM[column] < lower_bound) | (dfM[column] > upper_bound)].radius_mean.size}", desc[2])
```
