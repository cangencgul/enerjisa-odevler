```python
df = pd.read_csv("data.csv")
df.describe().T
```

Radius Mean Area Mean ve Compactnes Mean için IQR metoduyla outlier kalan veri sayısı

```python
columns = ["radius_mean", "area_mean", "compactness_mean"]
```

```python
for diagnosis in ["B", "M"]:
    dfB = df[df["diagnosis"] == diagnosis]
    print(f"DİAGNOSİS {diagnosis}")
    print(dfB[columns].describe())
    print(f"\nNumber Of Outlier value")
    for column in columns:
        desc = dfB[column].describe()
        Q1 = desc[4]
        Q3 = desc[6]
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        print(f"{column} için {lower_bound:.2f}, {upper_bound:.2f} aralığında " 
              f"{dfB[(dfB[column] < lower_bound) | (dfB[column] > upper_bound)].radius_mean.size} değer")
    print("\n")
```

2. soru için yorum

diagnosis B ve M için radius meanin mean ve median değerleri birbirine oldukça yakın ike compactness değeri için mean ve median değerleri arasında görece daha fazla fark vardır. Bu duruma paralel olarak outlier değer sayısı radius mean için düşük compacntess mean için ise yüksek çıkmıştır.

yani mean ile median arasındaki fark ile outlier değer arasında doğru orantı vardır.

```python
for diagnosis in ["B", "M"]:
    dfB = df[df["diagnosis"] == diagnosis]
    print(f"DİAGNOSİS {diagnosis}")
    kolum_ikili_combinasyonları = combinations(columns, 2)
    for i in list(kolum_ikili_combinasyonları):
        ortalamalar_farki = dfB[i[0]].mean() - dfB[i[1]].mean()  
        uzunluk_1 = len(i[0])
        uzunluk_2 = len(i[1])
        s1, s2 = dfB[i[0]].var(), dfB[i[1]].var()
        s = math.sqrt(((uzunluk_1 - 1) * s1 + (uzunluk_2 - 1) * s2) / (uzunluk_1 + uzunluk_2 - 2))
        effect_size = ortalamalar_farki / s
        print(i[0], i[1], f"için efekt size: {effect_size}")
```
